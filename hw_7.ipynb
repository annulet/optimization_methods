```python
import numpy as np
```

```python
parents = np.random.randint(50, size=(3, 4))
parents
```

```python
children_size = [3, 4]
```

```python
def one_point_crossever(parents, children_size):
    idx = np.random.randint(parents.shape[1], size=children_size[0])

    children = np.empty(children_size)

    for i in range(children.shape[0]):
        children[i] = np.concatenate((parents[i, idx[i]:], parents[i, : idx[i]]), axis=None)
    return children
```

```python
one_point_crossever(parents, children_size)
```
